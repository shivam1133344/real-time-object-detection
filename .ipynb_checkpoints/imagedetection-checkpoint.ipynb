```python
import cv2 
import matplotlib.pyplot as plt
from  ultralytics  import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
image= cv2.imread("test.jpg")
```

```python
result=model(image)
```

```python
annotated_frame=result[0].plot()
```

```python
plt.imshow(cv2.cvtColor(annotated_frame , cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
